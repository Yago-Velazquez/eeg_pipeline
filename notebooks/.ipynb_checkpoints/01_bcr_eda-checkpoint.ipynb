```python
# Cell 1 — Imports (paste as first cell)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (12, 4)
sns.set_style('darkgrid')

print("✅ Imports OK")
```

```python
# Cell 2 — Load BCR feature table
BCR_PATH = "data/raw/Bad_channels_for_ML.csv"

df = pd.read_csv(BCR_PATH)

print(f"Shape: {df.shape}")
print(f"\nColumns ({len(df.columns)}):")
print(df.columns.tolist())
print(f"\nFirst 3 rows:")
df.head(3)
```

```python
# Cell 3 — Parse Session into session_group, subject_id, visit_id
df['session_group'] = df['Session'].str.split('-').str[0].astype(int)
df['subject_id']    = df['Session'].str.split('-').str[1].astype(int)
df['visit_id']      = df['Session'].str.split('-').str[2].astype(int)

print("Unique session groups:", sorted(df['session_group'].unique()))
print("Unique subjects:      ", df['subject_id'].nunique())  # expect 43
print("Unique visits:        ", sorted(df['visit_id'].unique()))

# Check multi-site subjects (expect 6)
cross = df.groupby('subject_id')['session_group'].nunique()
multi_site = cross[cross > 1]
print(f"\nMulti-session-group subjects: {len(multi_site)}")  # expect 6
print(multi_site)

print(f"\nSample parsed rows:")
print(df[['Session', 'session_group', 'subject_id', 'visit_id']].head(5))
```

```python
# Cell 4 — impedance_missing feature
# Must be done BEFORE any imputation

# Check the exact column name for impedance in your file
impedance_cols = [c for c in df.columns if 'mpedance' in c or 'imp' in c.lower()]
print("Impedance columns found:", impedance_cols)

# Create missing flag from the primary impedance column
# Adjust 'Impedance_start' to the actual column name
df['impedance_missing'] = df['Impedance (start)'].isnull().astype(int)

missing_rate = df['impedance_missing'].mean()
print(f"\nimpedance_missing rate: {missing_rate:.1%}")  # expect ~31.3%
print(f"Sessions with missing: {df.groupby('Session')['impedance_missing'].max().sum()}")
# expect ~13 sessions

# Verify the flag correlates with sessions, not random rows
print("\nMissing by site:")
print(df.groupby('session_group')['impedance_missing'].mean().round(3))
```

```python
# Cell 5 — Bad_score distribution and binary target

# Score distribution
print("=== Bad_score distribution ===")
print(df['Bad (score)'].value_counts().sort_index())

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: score distribution
df['Bad (score)'].value_counts().sort_index().plot(kind='bar', ax=axes[0],
    color=['#4f9cf9' if i < 2 else '#f87171' for i in range(4)])
axes[0].set_title('Bad (score) distribution')
axes[0].set_xlabel('Score')
axes[0].axvline(1.5, color='orange', ls='--', label='threshold ≥2')
axes[0].legend()

# Binary target: score >= 2 = BAD
df['is_bad'] = (df['Bad (score)'] >= 2).astype(int)
bad_rate = df['is_bad'].mean()
scale_pos_weight = (1 - bad_rate) / bad_rate

print(f"\nBad rate (score≥2): {bad_rate:.3f} = {bad_rate:.1%}")
print(f"Good channels: {(df['is_bad']==0).sum():,}")
print(f"Bad channels:  {(df['is_bad']==1).sum():,}")
print(f"scale_pos_weight = {scale_pos_weight:.1f}  (use this in XGBoost)")

# Plot 2: binary class balance
df['is_bad'].value_counts().plot(kind='pie', ax=axes[1],
    labels=['Good', 'Bad'], autopct='%1.1f%%',
    colors=['#4f9cf9', '#f87171'])
axes[1].set_title(f'Class balance (bad rate = {bad_rate:.1%})')

# Plot 3: bad rate by site
site_bad = df.groupby('session_group')['is_bad'].mean().sort_values(ascending=False)
site_bad.plot(kind='bar', ax=axes[2], color='#f6ad55')
axes[2].set_title('Bad rate per site')
axes[2].set_ylabel('Bad channel rate')
axes[2].axhline(bad_rate, color='red', ls='--', label='overall mean')
axes[2].legend()

plt.tight_layout()
plt.savefig('results/figures/bcr_class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Cell 6 — Inter-rater correlation (direct, using Bad (site X) columns)
rater_cols = ['Bad (site 2)', 'Bad (site 3)', 'Bad (site 4a)', 'Bad (site 4b)']

# Correlation matrix
corr = df[rater_cols].corr()
print("Inter-rater correlation matrix:")
print(corr.round(3))

# Plot
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt='.3f', cmap='RdYlGn',
            vmin=-1, vmax=1, ax=ax)
ax.set_title('Inter-site rater correlation')
plt.tight_layout()
plt.savefig('results/figures/bcr_rater_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

# Summary: site 3 mean correlation with all others
site3_mean_corr = corr['Bad (site 3)'].drop('Bad (site 3)').mean()
print(f"\nSite 3 mean correlation with all other sites: {site3_mean_corr:.3f}")
print("Interpretation: this is label noise, not model failure.")
```

```python
# Cell 6b — Confirm target variable and document threshold decision
# is_bad was created in Cell 5, just verify it's correct
print("Target variable summary:")
print(f"  Column used: 'Bad (score)' >= 2")
print(f"  Bad channels:  {df['is_bad'].sum():,}")
print(f"  Good channels: {(df['is_bad']==0).sum():,}")
print(f"  Bad rate:      {df['is_bad'].mean():.3f}")
print(f"  scale_pos_weight: {(df['is_bad']==0).sum() / df['is_bad'].sum():.1f}")

# Score=1 borderline zone — flagged for Day 10 threshold sensitivity
borderline = (df['Bad (score)'] == 1).sum()
print(f"\nBorderline (score=1): {borderline:,} rows ({borderline/len(df):.1%})")
print("These are treated as GOOD under threshold>=2 but will be")
print("tested in Day 10 threshold sensitivity experiment.")

# GroupKFold groups array — save this, you'll use it in Week 2
groups = df['subject_id'].values
print(f"\nGroupKFold groups array ready: {len(np.unique(groups))} unique subjects")
```

```python
# Cell 7 — Bad rate by channel label
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Plot 1: bad rate per channel (sorted descending)
if 'Channel labels' in df.columns:
    ch_bad = df.groupby('Channel labels')['is_bad'].mean().sort_values(ascending=False)
    ch_bad.plot(kind='bar', ax=axes[0,0], color='#f87171', alpha=0.8)
    axes[0,0].set_title('Bad rate per channel (sorted)')
    axes[0,0].set_ylabel('Bad channel rate')
    axes[0,0].axhline(bad_rate, color='blue', ls='--', label='mean')
    axes[0,0].tick_params(axis='x', rotation=90)
    axes[0,0].legend()
    # Print top 5 worst channels
    print("Top 5 worst channels:")
    print(ch_bad.head(5))
    # Expect: T7, T8, AF7, M1, M2 to be top offenders

# Plot 2: bad rate per subject
subj_bad = df.groupby('subject_id')['is_bad'].mean().sort_values(ascending=False)
subj_bad.plot(kind='bar', ax=axes[0,1], color='#a78bfa', alpha=0.8)
axes[0,1].set_title('Bad rate per subject (sorted)')
axes[0,1].set_ylabel('Bad channel rate')
axes[0,1].axhline(bad_rate, color='red', ls='--')
axes[0,1].tick_params(axis='x', rotation=90)

# Plot 3: impedance_missing by site
imp_site = df.groupby('session_group')['impedance_missing'].mean()
imp_site.plot(kind='bar', ax=axes[1,0], color='#fbbf24')
axes[1,0].set_title('Impedance missingness rate by site')
axes[1,0].set_ylabel('% missing')

# Plot 4: bad rate per task (if task column exists)
task_cols = [c for c in df.columns if 'task' in c.lower()]
if task_cols:
    task_bad = df.groupby(task_cols[0])['is_bad'].mean().sort_values(ascending=False)
    task_bad.plot(kind='bar', ax=axes[1,1], color='#34d399')
    axes[1,1].set_title(f'Bad rate per {task_cols[0]}')
else:
    axes[1,1].text(0.5, 0.5, 'No task column found',
                    ha='center', va='center', transform=axes[1,1].transAxes)

plt.tight_layout()
plt.savefig('results/figures/bcr_eda_overview.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Cell 8 — Verify GroupKFold by subject (no leakage check)
from sklearn.model_selection import GroupKFold

groups = df['subject_id'].values
X_dummy = np.zeros((len(df), 1))
y_dummy = df['is_bad'].values

gkf = GroupKFold(n_splits=5)

print("=== GroupKFold(5) by subject_id ===")
for fold, (train_idx, val_idx) in enumerate(gkf.split(X_dummy, y_dummy, groups)):
    train_subjs = set(groups[train_idx])
    val_subjs   = set(groups[val_idx])
    overlap     = train_subjs & val_subjs
    val_bad     = y_dummy[val_idx].mean()

    print(f"Fold {fold+1}: train={len(train_idx):,} val={len(val_idx):,} | "
          f"val_subjects={len(val_subjs)} | overlap={len(overlap)} | val_bad_rate={val_bad:.3f}")

# overlap should be 0 for ALL folds — if not, your grouping is wrong
```
